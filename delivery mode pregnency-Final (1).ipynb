##### Performing encoding

```python
df_dummies = pd.get_dummies(df,columns = ['Ob_Score'], drop_first= True)
df_dummies.shape
```

###### splitting dataset into X and Y variable

```python
x= df_dummies.drop(['Del_mode'],axis = 1)
y= df_dummies["Del_mode"]
```

```python
x.shape, y.shape
```

```python
x
```

```python
y.value_counts()
```

###### Balancing the output by performing oversampling

```python
from imblearn.over_sampling import RandomOverSampler
ros= RandomOverSampler()
x_resampled , y_resampled = ros.fit_resample(x,y)
```

###### performing Chi square test for feature selection

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
select_kbest_rank_features= SelectKBest(score_func= chi2, k=5)
kbest_features= select_kbest_rank_features.fit(x_resampled,y_resampled)

df_kscore= pd.DataFrame(kbest_features.scores_,columns=['Score'])
df_columns= pd.DataFrame(x_resampled.columns)

kbest_features_rank= pd.concat([df_kscore,df_columns], axis=1)
kbest_features_rank.columns= ['k_score', 'features']

kbest_features_rank
print(kbest_features_rank.sort_values(by='k_score', ascending= False))
```

```python
X_resampled = x_resampled.copy()
```

```python
X_resampled1 = X_resampled.drop(columns = ['Cerv_Len_cms','Dialation_Score','Consistency_Score'])
```

##### spliting data into training and testing

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size = 0.2, random_state = 0)
```

##### data pre processing

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler().fit(x_train)
scaler = RobustScaler().fit(x_test)
```

```python
scaler.n_features_in_
```

```python
X_train_scaled = scaler.transform(x_train)
X_test_scaled = scaler.transform(x_test)
```

# Model building

### Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
classifier1 = DecisionTreeClassifier()
classifier1.fit(X_train_scaled,y_train)
```

```python
y_pred1 = classifier1.predict(X_test_scaled)
```

confusion matrix

```python
from sklearn.metrics import confusion_matrix, classification_report,accuracy_score, precision_score, recall_score
```

```python
print(confusion_matrix(y_pred1,y_test))
print('accuracy % =', accuracy_score(y_test,y_pred1))
print(classification_report(y_test,y_pred1))
print('precision score =', precision_score(y_test,y_pred1))
print('recall_score =', recall_score(y_test,y_pred1))
```

### Random forest

```python
from sklearn.ensemble import RandomForestClassifier
classifier2 = RandomForestClassifier()
classifier2.fit(X_train_scaled, y_train)
```

```python
y_pred2 = classifier2.predict(X_test_scaled)
```

```python
print(confusion_matrix(y_pred2,y_test))
print('accuracy % =', accuracy_score(y_test,y_pred2))
print(classification_report(y_test,y_pred2))
print('precision score =', precision_score(y_test,y_pred2))
print('recall_score =', recall_score(y_test,y_pred2))
```

```python
DT = classifier1.fit(X_train_scaled, y_train)
probs_DT = DT.predict_proba(X_test_scaled)[:, 1]

RF = classifier2.fit(X_train_scaled, y_train)
probs_RF = RF.predict_proba(X_test_scaled)[:, 1]
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
```

```python
auc_dt = roc_auc_score(y_test, probs_DT)
fpr_dt, tpr_dt, thresholds_lr = roc_curve(y_test, probs_DT)

auc_rf = roc_auc_score(y_test, probs_RF)
fpr_rf, tpr_rf, thresholds_lr = roc_curve(y_test, probs_RF)
```

```python
plt.figure(figsize=(12, 7))
plt.plot(fpr_dt, tpr_dt, label=f'AUC (Decision Tree) = {auc_dt:.2f}')
plt.plot(fpr_rf, tpr_rf, label=f'AUC (Random Forests) = {auc_rf:.2f}')
plt.plot([0, 1], [0, 1], color='blue', linestyle='--', label='Baseline')
plt.title('ROC Curve', size=20)
plt.xlabel('False Positive Rate', size=14)
plt.ylabel('True Positive Rate', size=14)
plt.legend();
```
